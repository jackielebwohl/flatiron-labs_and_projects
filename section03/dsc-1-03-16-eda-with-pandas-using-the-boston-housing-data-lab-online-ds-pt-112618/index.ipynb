```python
fourth_quartile=df.loc[df['crim'] > 3.68]
fourth_quartile.head()

first_quartile=df.loc[df['crim'] < .079]
first_quartile.tail()

fourth_quartile.tail(50)

```

```python
fourth_quartile.describe()
```

```python
first_quartile.describe()
```

```python
plt.hist(first_quartile['age'], color='blue', label='Least Crime')
plt.hist(fourth_quartile['age'], color='red', label='Most Crime')
plt.title('Age of Home and Crime Rate')
plt.legend()
plt.xlabel('Age of Home')
plt.ylabel('Number of Homes')
plt.show()
```

Crime and Property Value 

```python
plt.scatter(first_quartile['medv'], first_quartile['crim'], color='blue', label='Least Crime')
plt.scatter(fourth_quartile['medv'],fourth_quartile['crim'], color='red', label='Most Crime')
plt.title('Value of Home vs. Per Capita Crime Rate')
plt.legend()
plt.xlabel('Median Home Value in $10,000s')
plt.ylabel('Crime Rate')
plt.show()
```

```python
plt.scatter(first_quartile['lstat'], first_quartile['crim'], color='blue', label='Least Crime')
plt.scatter(fourth_quartile['lstat'],fourth_quartile['crim'], color='red', label='Most Crime')
plt.title('Socio-Economic Status vs. Per Capita Crime Rate')
plt.legend()
plt.xlabel('Percentage of "Lower Status" Population')
plt.ylabel('Crime Rate')
plt.show()
```

**Properties Adjacent to Charles

```python
not_adjacent=df.loc[df['chas'] == 0]
adjacent=df.loc[df['chas']==1]
adjacent.describe()
```

```python
not_adjacent.describe()
```

Median Home Value

```python
new_figure = plt.figure(figsize=(10,4))

# Add a subplot to the figure - a new axes
ax = new_figure.add_subplot(121)

# Add a second subplot to the figure - a new axes
ax2 = new_figure.add_subplot(122)

# Generate a line plot on first axes
ax.hist(not_adjacent['medv'], color='lightblue', bins=5)

# Draw a scatter plot on 2nd axes
ax2.hist(adjacent['medv'], color='magenta', bins=5)


ax.set_title('Median Value Homes Not Adjacent to Charles River')
ax2.set_title('Median Value Homes Adjacent to Charles River')

ax.set_xlabel('Home Value in $10,000s')
ax2.set_xlabel('Home Value in $10,000s')
ax.set_ylabel('Number of Towns')

# Show the plot
plt.show()
```

Age of Home 

The result surprised me here. While looking at the .details() report I would have assumed that the homes adjacent to the Charles are generally older. It is clear looking at this chart that the age seems to follow similar trends all homes in the region. However, it is worth noting that there is an extremely small number of homes adjacent to the Charles perhaps rendering these findings insignificant. 

```python
plt.hist(not_adjacent['age'], color='lightblue', label='Not Adjacent to Charles')
plt.hist(adjacent['age'], color='magenta', label='Adjacent to Charles')
plt.title('Age of Home and Adjacence to Charles River')
plt.legend()
plt.xlabel('Age of Home')
plt.ylabel('Number of Homes')
plt.show()
```

Percentage industrialized

```python
plt.hist(not_adjacent['indus'], color='lightblue', label='Not Adjacent to Charles')
plt.hist(adjacent['indus'], color='magenta', label='Adjacent to Charles')
plt.title('Percentage of Industrial Land')
plt.legend()
plt.xlabel('Percentage of Industrial Land')

plt.show()
```

**Number of Rooms in Home**

```python
most_rooms=df.loc[df['rm'] > 6.595]
most_rooms.head()

fewest_rooms=df.loc[df['rm'] < 5.884]
fewest_rooms.describe()
```

```python
most_rooms.describe()
```

Crime Rates

```python




new_figure = plt.figure(figsize=(10,4))

# Add a subplot to the figure - a new axes
ax = new_figure.add_subplot(121)

# Add a second subplot to the figure - a new axes
ax2 = new_figure.add_subplot(122)

# Generate a line plot on first axes
ax.hist(most_rooms['crim'], color='green', bins=10, label='Most Rooms')

# Draw a scatter plot on 2nd axes
ax2.hist(fewest_rooms['crim'], color='orange', bins=10, label='Fewest Rooms')


ax.set_title('Crime Rate per Capita in Homes with Most Rooms')
ax2.set_title('Crime Rate per Capita in Homes with Fewest Rooms')

ax.set_xlabel('Crime Rate')
ax2.set_xlabel('Crime Rate')
ax.set_ylabel('Number of Homes')

ax.set_xlim(0, 50)

# Show the plot
plt.show()
```

# Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centres.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.
  
  
  
Source
Harrison, D. and Rubinfeld, D.L. (1978) Hedonic prices and the demand for clean air. J. Environ. Economics and Management 5, 81102.

Belsley D.A., Kuh, E. and Welsch, R.E. (1980) Regression Diagnostics. Identifying Influential Data and Sources of Collinearity. New York: Wiley.
